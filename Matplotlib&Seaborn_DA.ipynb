# SCATTER PLOT

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 63, 70, 75, 78, 85]
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")

plt.scatter(study_hours,exam_scores,marker='o')
plt.show()
```

```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np

data = {
    "total_bill": np.round(np.random.uniform(10, 60, 30), 2),
    "tip": np.round(np.random.uniform(2, 10, 30), 2),
    "sex": np.random.choice(["Male", "Female"], 30),
    "smoker": np.random.choice(["Yes", "No"], 30),
    "day": np.random.choice(["Thur", "Fri", "Sat", "Sun"], 30),
    "time": np.random.choice(["Lunch", "Dinner"], 30),
    "size": np.random.randint(1, 6, 30)
}

df = pd.DataFrame(data)
print(df)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=data)
plt.title("Total Bill v/s Tip")
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",data=data)
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=data)
plt.title("Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=data)
plt.show()
```

```python
sns.histplot(data["total_bill"],bins=5,kde=True)
plt.show()
```

```python
sns.kdeplot(data["total_bill"],fill=True)
plt.show()
```
